```python
import pandas as pd
```

### Import the full survey

```python
file_path = "Y:/GitHub/SentientSpectators/annotation_evaluation/survey2-4annotators_full.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
df
```

### Separate the first 6 columns (information on annotators) and the rest (question columns)

```python
info_df = df.iloc[:, :6]  
annotator_df = df.iloc[:, 6:]  
```

```python
info_df
```

###  Separate the columns where we have 3 or 4 (out of 4) Yes'

```python
# counts the number of 'Yes' responses in the column
# returns True if the count is 3 or 4,else False 

def filter_columns(column):
    yes_count = (column == 'Yes').sum()
    return yes_count == 3 or yes_count == 4
```

```python
filtered_columns = annotator_df.apply(filter_columns)
```

```python
filtered_annotator_df = annotator_df.loc[:, filtered_columns]
```

```python
filtered_annotator_df
```

```python
filtered_annotator_df = filtered_annotator_df.rename(columns=lambda x: x.replace('[sic]', ''))
filtered_annotator_df.shape
```

```python
sentences = []
annotations = []
```

```python
for column in filtered_annotator_df.columns:
    sentence, annotation = column.split('[', 1)
    sentences.append(sentence.strip())
    annotations.append(annotation)
```

```python
len(sentences)
```

```python
sentences
```

### create the annotation and sentence columns, by getting rid of superflous strings

```python
new_df = pd.DataFrame({
    'sentence': sentences,
    'annotation': annotations
}) 
```

```python
with pd.option_context('display.max_rows', None,):
    print(new_df)
```

```python
# Clean the 'annotation' column
new_df['annotation'] = new_df['annotation'].str.replace(r'\s*\].*', '', regex=True)
```

```python
new_df['sentence'] = new_df['sentence'].str.extract(r'(?:Previous sentence:.*?)*?To annotate:?\s*(.*?)(?:\s*Following sentence:.*|$)', expand=False)
```

```python
unique_values = new_df['annotation'].unique()

# To print the unique values
print(unique_values)
```

```python
with pd.option_context('display.max_rows', None,):
    print(new_df)
```

### Separate polarity and emotion annotations

```python
polarities = ['positive', 'negative', 'neutral']
emotions = ['joy', 'trust', 'anticipation', 'surprise', 'anger', 'disgust', 'sadness', 'fear', 'no emotion']
```

```python
df_polarities = new_df[new_df['annotation'].isin(polarities)]
df_emotions = new_df[new_df['annotation'].isin(emotions)]
```

```python
df_polarities.shape
```

```python
df_polarities
```

```python
df_emotions.shape
```

### Saving both dataframes as csvs

```python
df_polarities.to_csv('survey2_gold_polarities.csv', index=False, encoding='utf-8')
```

```python
df_emotions.to_csv('survey2_gold_emotions.csv', index=False, encoding='utf-8')
```

### Timing analysis (survey1)

```python
file_path = "Y:/GitHub/SentientSpectators/annotation_evaluation/survey1-time.csv"
```

```python
df = pd.read_csv(file_path, encoding = 'latin1')
```

```python
df
```

```python
def seconds_to_hms(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = int(seconds % 60)
    return f"{hours:02d}:{minutes:02d}:{seconds:02d}"
```

```python
time_df = df.iloc[:, 6:].applymap(seconds_to_hms)
```

```python
time_df
```

Highest time was by Yvonne 03:23 hours, then Elisabeth 2:57, then Ertler 1:56 then Pascal with 54 minutes (first survey)

```python
question_total_time_seconds = df.iloc[:, 6:].sum(axis=0)
top_5_questions_seconds = question_total_time_seconds.nlargest(5)
top_5_questions_hours_minutes = top_5_questions_seconds.apply(seconds_to_hms)
```

```python
print(top_5_questions_hours_minutes)
```

- Total time                09:11:59
- Question time: G01Q27     00:17:15
- Question time: G00Q101    00:14:41
- Question time: G00Q103    00:13:29
- Question time: G00Q107    00:11:16
