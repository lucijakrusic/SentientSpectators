###  Separate the columns where we have 3 or 4 (out of 4) Yes'

In order to get the majority vote for both polarities and emotions 

```python
# counts the number of 'Yes' responses in the column
# returns True if the count is 3 or 4,else False 

def filter_columns(column):
    yes_count = (column == 'Yes').sum()
    return yes_count == 3 or yes_count == 4
```

### Creating a dataframe with following and previous sentence:

```python
data = []
```

```python
def extract_parts_from_column(column_name):
    # Step 1: Extract the question code (before the first bracket or any punctuation)
    question_code_match = re.search(r"([A-Za-z0-9]+Q\d+)", column_name)
    question_code = question_code_match.group(1).strip() if question_code_match else None

    # Step 2: Extract the sentence following "To annotate:"
    sentence_match = re.search(r"To annotate:? ?(.*?)(?= \[| Following sentence|$)", column_name, re.DOTALL)
    to_annotate = sentence_match.group(1).strip() if sentence_match else None

    # Step 3: Extract the previous sentence (if it exists)
    previous_sentence_match = re.search(r"Previous sentence:? ?(.*?)(?= To annotate|$)", column_name, re.DOTALL)
    previous_sentence = previous_sentence_match.group(1).strip() if previous_sentence_match else "-"

    # Step 4: Extract the following sentence (if it exists)
    following_sentence_match = re.search(r"Following sentence:? ?(.*?)(?= \[|$)", column_name, re.DOTALL)
    following_sentence = following_sentence_match.group(1).strip() if following_sentence_match else "-"

    # Step 5: Extract the annotation (text after the last occurrence of '[')
    annotation = column_name.rsplit('[', 1)[-1].replace(']', '').strip() if '[' in column_name else None

    return question_code, previous_sentence, to_annotate, following_sentence, annotation

# Loop through the columns and extract all parts
for column in columns:
    question_code, previous_sentence, to_annotate, following_sentence, annotation = extract_parts_from_column(column)
    # Include partial matches as well
    if question_code or to_annotate or annotation:
        data.append([question_code, previous_sentence, to_annotate, following_sentence, annotation])

```

```python
df_new = pd.DataFrame(data, columns=['question_code', 'previous_sentence', 'to_annotate', 'following_sentence', 'annotation'])
```

### create a new df with three columns - question_code, sentence and annotation (without the previous and following sentence):

data wrangling necessary due to output format by LimeSurvey

```python
# Function to handle extraction from a column name
def extract_parts_after_last_bracket(column_name):
    # Step 1: Extract the question code (before the first bracket or any punctuation)
    question_code_match = re.search(r"([A-Za-z0-9]+Q\d+)", column_name)
    question_code = question_code_match.group(1).strip() if question_code_match else None

    # Step 2: Extract the sentence following "To annotate:"
    sentence_match = re.search(r"To annotate:? ?(.*?)(?= \[| Following sentence|$)", column_name, re.DOTALL)
    sentence = sentence_match.group(1).strip() if sentence_match else None

    # Step 3: Extract the annotation (text after the last occurrence of '[')
    annotation = column_name.rsplit('[', 1)[-1].replace(']', '').strip() if '[' in column_name else None

    return question_code, sentence, annotation
```

```python
data = []
```

```python
for column in filtered_annotator_df.columns:
    question_code, sentence, annotation = extract_parts_after_last_bracket(column)
    # Include partial matches as well
    if question_code or sentence or annotation:
        data.append([question_code, sentence, annotation])

```

### Timing analysis (survey1)

```python
file_path = "D:/GitHub/SentientSpectators/1_sentiment_annotation/input/3_round/survey1-time.csv"
```

```python
df = pd.read_csv(file_path, encoding = 'latin1')
```

```python
df
```

```python
def seconds_to_hms(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = int(seconds % 60)
    return f"{hours:02d}:{minutes:02d}:{seconds:02d}"
```

```python
time_df = df.iloc[:, 6:].applymap(seconds_to_hms)
```

```python
time_df
```

Highest time was by annotator1 03:23 hours, then annotator2 2:57, then annotator3 1:56 then annotator4 with 54 minutes (first survey)

```python
question_total_time_seconds = df.iloc[:, 6:].sum(axis=0)
top_5_questions_seconds = question_total_time_seconds.nlargest(5)
top_5_questions_hours_minutes = top_5_questions_seconds.apply(seconds_to_hms)
```

```python
print(top_5_questions_hours_minutes)
```
